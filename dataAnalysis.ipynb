```python
import pandas
import numpy
import seaborn as sn
import matplotlib.pyplot as plt
```

```python
df=pandas.read_csv("titanic_cleaned.csv")
df.head()
```

```python
#kde-kernel density estimate = to draw a smooth curve on the histogram plot
sn.histplot(df['Age'],kde=True)
plt.title("Distribution of Age")
```

```python
sn.histplot(df['Fare'],kde=True)
plt.title("Distribution of Fare")
```

```python
sn.boxplot(x=df['Age'])
plt.title("Boxplot of Age")
```

```python
sn.boxplot(x=df['Fare'])
plt.title("Boxplot of Fare")
```

```python
sn.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title("Correlation Matrix Heatmap")
```

```python
sn.countplot(x='Sex', data=df)
plt.title("Count of Passengers by Gender")
```

```python
sn.countplot(x='Survived', data=df)
plt.title("Survided vs Died")
```

```python
sn.countplot(x='Sex', hue='Survived', data=df)
```
